```python
data_Dep_Time['Fly_Hours'] = 0
for i in range(0,len(data_Dep_Time['Hours'])):
    val = data_Dep_Time.loc[i, 'Hours']
    if val == 0:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Midnight'
    elif val >= 1 and val < 4:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Early Morning'
    elif val >= 4 and val < 6:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Dawn'
    elif val >= 6 and val < 9:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Morning'
    elif val >= 9 and val < 12:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Mid-Morning'
    elif val >= 12 and val < 13:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Noon'
    elif val >= 13 and val < 16:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'After Noon'
    elif val >= 16 and val < 21:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Evening'
    elif val >= 21 and val < 23:
        data_Dep_Time.loc[i,'Fly_Hours'] = 'Night'
```
